# 4. 데이터 전처리하기

```python
df['상호명'].value_counts()
```

```python
starbucks = []

for i in range(df.shape[0]) :
    if '스타벅스' in df['상호명'][i] :
        starbucks.append(df['상호명'][i])
```

```python
starbucks
```

```python
def unite_cafe_name(x) :
    if '스타벅스' in x :
        return '스타벅스'
    elif '커피빈' in x :
        return '커피빈'
    elif '할리스' in x :
        return '할리스'
    elif '투썸' in x :
        return '투썸플레이스'
    elif '이디야' in x :
        return '이디야'
    elif '공차' in x :
        return '공차'
    elif '카페베네' in x :
        return '카페베네'
    elif '엔제리너스' in x :
        return '엔제리너스'
    elif '탐앤탐스' in x :
        return '엔제리너스'
    elif '빽다방' in x :
        return '빽다방'
    
    else :
        return x
```

```python
df.loc[:, '상호명'] = df['상호명'].apply(lambda x : unite_cafe_name(x))
```

# 6. 서울 지도에 카페 Top5 카페 점포수 시각화 하기

```python
df_top20['상호명'][1:6]
```

```python
def etc_cafe_name(x) :
    if x in list(df_top20['상호명'][1:6]) :
        return x
    else :
        return '기타_카페'
```

```python
df.loc[:, '상호명'] = df['상호명'].apply(lambda x : etc_cafe_name(x))
```

```python
df = df[df['상호명'] != '기타_카페']
```

```python
df.head()
```

```python
df['상호명'].value_counts()
```

```python
def cafe_color(x) :
    if x == '스타벅스' :
        return 'green'
    elif x == '이디야' :
        return 'blue'
    elif x == '투썸플레이스' :
        return 'red'
    elif x == '엔제리너스' :
        return 'orange'
    elif x == '커피빈' :
        return 'beige'
```

```python
df['표시색상'] = df['상호명']
df.head()
```

```python
df.loc[:, '표시색상'] = df['표시색상'].apply(lambda x : cafe_color(x))
```

```python
df.reset_index(inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
cafe_map = folium.Map(location=[df['위도'].median(), df['경도'].median()], zoom_start = 12)

for i in range(1400) :
    
    folium.Marker(location =  [df.loc[i, '위도'], df.loc[i, '경도']],
                  popup = df['상호명'][i],
                  icon = folium.Icon(color = df['표시색상'][i], icon = 'glass')).add_to(cafe_map)

cafe_map
# cafe_map.save('seoul_cafe_map.html')
```
